# Preparing data

## Reading data

```python
import pandas as pd
df = pd.read_csv('data/sentiment.csv')
df.head(20)
```

## Chinese Tokenization with jieba

```python
import jieba
df['token_text'] = df['text'].apply(lambda x:list(jieba.cut(x)))
df.head(20)
```

## Loading stop words

```python
with open("data/stopwords_zh-tw.txt", encoding="utf-8") as fin:
    stopwords = fin.read().split("\n")[1:]
print(stopwords[:100])
print(len(stopwords))
```

```python
def remove_stopWords(words):
    out = []
    for word in words:
        if word not in stopwords:
            out.append(word)
    return out


import unicodedata # for removing Chinese puctuation
def remove_punc_by_unicode(words):
    out = []
    for word in words:
        if word != " " and not unicodedata.category(word[0]).startswith('P'):
            out.append(word)
    return out
```

```python
df['cleaned'] = df['token_text'].apply(remove_stopWords)
df['cleaned'] = df['cleaned'].apply(remove_punc_by_unicode)
df
```

# Feature selection

## tfidf vectorization
1. https://towardsdatascience.com/clustering-documents-with-python-97314ad6a78d
2. https://blog.csdn.net/blmoistawinde/article/details/80816179

```python
doc = ['hahaha', 'heheh', 'testing']
" ".join(doc)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
documents = [" ".join(doc) for doc in df['cleaned']]
documents[:10]
```

### Method 1. 

```python
vectorizer = TfidfVectorizer(
    'ngram_range': (1, 2),
    'max_df': 0.2,
    'token_pattern': r"(?u)\b\w+\b",
    'use_idf': True
)
X = vectorizer.fit_transform(documents)
len(documents)
```

### Model 1. 
去掉停用詞後做tf-idf

```python
# tf-idf 只保留兩個字以上的 term

tfidf_model = TfidfVectorizer().fit(documents)
sparse_result = tfidf_model.transform(documents)
print(sparse_result) # (document_id, term_code): vectorized point
print(len(tfidf_model.vocabulary_)) # tfidf_model.vocabulary_ 算 documents的 term frequency
```

### Model 2
用regex把一個字的term也留下來

```python
tfidf_model2 = TfidfVectorizer(token_pattern=r"(?u)\b\w+\b").fit(documents)
print(len(tfidf_model2.vocabulary_))
```

### Model 3

- 當然有些時候我們還是要過濾掉一些無意義的詞，留下一些不是每份文件都有的字。也就是說，我們希望某些字只有`不超過某個比例`的文件擁有，可以透過調整參數調整這些字詞的權重來實現這一目的：`max_df(document frequency): [0.0, 1.0]`內的浮點數或正整數, 默認值=`1.0`。當設置為浮點數時，會留下只有該浮點數比例以內的文章才擁有的詞語，也就是說max_df參數越小，留下來的詞語的在各文件之間的獨特性越高。可以搭配這個方法與停用詞一起將無用或太常見的字詞一起剔除。這樣就可以幫助我們過濾掉出現太多的無意義詞語，如下面的"我"就被過濾（雖然這里“我”的排比在文學上是很重要的）。
- 有11860個字是其他99%的文件所沒有的，與model 1相比，用max_df只多過濾104個字詞，代表這個tf-idf是個稀疏矩陣。拿去做clustering會很難做。

```python
# token_pattern=r"(?u)\b\w+\b", 
tfidf_model3 = TfidfVectorizer(max_df=0.01).fit(documents)  
print(len(tfidf_model3.vocabulary_))
```

### Model 4 filter out stop words

```python
tfidf_model4 = TfidfVectorizer(max_df=0.01,
                               # token_pattern=r"(?u)\b\w+\b", 
                               # max_features = 2000, # 指定要留下 tf-idf vector前 2000大的字詞，但不太建議。因為有些字的 vectorized point本來就比較小，很容易刪過頭。
                               stop_words=stopwords).fit(documents)
print(len(tfidf_model4.vocabulary_))
```

### Model 5 Only focus on some vocabs

```python
words = ['給力', '不錯', '不太好', '相當快', '無趣', '還不錯']
word_dict = {w:i for i, w in enumerate(words)} # i是 index，w是 words裡面的物件

# for i,w in enumerate(words):
#     word_dict[w] = i
print(word_dict)
tfidf_model5 = TfidfVectorizer(token_pattern=r"(?u)\b\w+\b",
                               vocabulary=word_dict).fit(documents) #只考慮 word_dict的字(已 mapping到 index上)
print(tfidf_model5.vocabulary_)
print(tfidf_model5.transform(documents).todense())
```

### Model 6 bigrams

```python
tfidf_model6 = TfidfVectorizer(ngram_range=(1,2), # 以 1-2個字詞為一組
                               max_df=0.01).fit(documents)
print(len(tfidf_model6.vocabulary_))
```

### Model 6 Bigrams limited max term number

```python
tfidf_model6 = TfidfVectorizer(token_pattern = r"(?u)\b\w+\b", 
                               max_features = 2000, 
                               ngram_range = (1,2),
                               stop_words = stopwords).fit(documents)
print(len(tfidf_model6.vocabulary_))
```

# Clustering

## Evaluation

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

Sum_of_squared_distances = []
K = range(2,20)

for k in K:
    km = KMeans(
        n_clusters=k, # 預設有 k個起始點做為 k個 clusters的 centroids
        max_iter=200, # iteration要重複幾次才能讓 squared point-centroid distance converged (1次 iteration會計算哪些資料點與起始 centroids距離最近，並依據這些資料點更新 centorids位置)
        n_init=10 # 對於每一個 k，要從新初始化幾次
    )
    km = km.fit(X)
    Sum_of_squared_distances.append(km.inertia_)

# 如果曲線不夠平緩就代表 clustering的效果不好    
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
def generate_color(x, total):
    def color_red_blue(ctr, reverse=False):
        r = int(ctr*255)
        b = 255-int(ctr*255)
        if reverse:
            return "#%02x%02x%02x" % (b, 0, r)
        else:
            return "#%02x%02x%02x" % (r, 0, b)
    return color_red_blue(x/total)

print(generate_color(1, 6))
```

## Store to dict+list for bar chart


- Using defaultdict: https://docs.python.org/3/library/collections.html

```python
from collections import defaultdict, Counter

cdict = defaultdict(Counter) # 是一個兩層的 dict結構，第一層是 dict，第二層是 Counter

for k, v in zip(df['cluster'], df['cleaned']):
    for word in v:
        cdict[k][word] += 1
for c in cdict:
    print(cdict[c].most_common(20))
    print("-"*80)
```
