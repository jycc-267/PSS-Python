### Q01.5 寫一函式去除所有中文停用詞`removeChiStopWords()`
```
def removeChiStopWords(...):
    ...

filtered = removeChiStopWords(filtered)
print(filtered)
```

````python
def removeChiStopWords(tf):
  token_clean = {k:v for k,v in tf.items() if k not in stopwords}
  return(token_clean)

filtered = removeChiStopWords(filtered)
# print(filtered)
print(len(filtered))
````

### Q01.6 寫一函式去除所有的英文字母與數字`removeAlphaNum()`。
```
def removeAlphaNum(...):
    ...
filtered = removeAlphaNum(filtered)
print(filtered)
```

````python
# YOUR CODE SHOULD BE HERE
# YOU CAN CHANGE THE VARIABLE NAME
def removeAlphaNum(tf):
  import re
  token_clean = {k:v for k,v in tf.items() if re.sub('[a-zA-Z]+|[0-9.]+', "", k) != ""}
  return(token_clean)

filtered = removeAlphaNum(filtered)
# print(filtered)
print(len(filtered))
````

### Q01.7 寫一函式去除全半形標點符號或空白：`removalPuncSpace()`。
```
def removalPuncSpace(...):
    ...
filtered = removalPuncSpaces(filtered)
print(filtered)
```

````python
def removealPuncSpace(tf):
  token_clean = {k:v for k,v in tf.items() if k not in punctuation and k.isspace()==False and k.isalnum()}
  return(token_clean)

filtered = removealPuncSpace(filtered)
# print(filtered)
print(len(filtered))
````
